```python
# Install necessary libraries if not already installed
# !pip install pandas

# Import libraries
import pandas as pd
import re

# Load the dataset
# Replace 'your_file.csv' with the actual file path or upload the file manually in Colab
df = pd.read_csv('/content/scopus_waam AND machine AND learning.csv')

# Extract the relevant columns (Title and Abstract)
abstracts = df[['Title', 'Abstract Note']].dropna()

# Define keywords for ethical and legal aspects
keywords = ['ethic', 'ethical', 'legal', 'compliance', 'regulation', 'law', 'governance']

# Function to highlight keywords in abstracts
def highlight_keywords(text, keywords):
    if pd.isna(text):
        return ""
    # Highlight each keyword using uppercase (or customize it)
    for keyword in keywords:
        text = re.sub(f"\\b{keyword}\\b", f"**{keyword.upper()}**", text, flags=re.IGNORECASE)
    return text

# Apply keyword highlighting to abstracts
abstracts['Abstract Note Highlighted'] = abstracts['Abstract Note'].apply(
    lambda x: highlight_keywords(x, keywords)
)

# Filter abstracts containing keywords
ethical_legal_abstracts = abstracts[
    abstracts['Abstract Note'].str.contains('|'.join(keywords), case=False, na=False)
]

# Display the filtered abstracts with highlighted keywords
print("Abstracts Related to Ethical and Legal Aspects (Highlighted Keywords):")
display(ethical_legal_abstracts[['Title', 'Abstract Note Highlighted']])

# Optional: Save the filtered abstracts with highlights to a new CSV file
ethical_legal_abstracts.to_csv('ethical_legal_abstracts_highlighted_ML.csv', index=False)
```

```python
# Install necessary libraries (if not already installed)
# !pip install pandas

# Import libraries
import pandas as pd

# Load the dataset
# Replace 'your_file.csv' with the actual file path or upload the file manually in Colab
df = pd.read_csv('/content/scopus_waam AND machine AND learning.csv')

# Extract the relevant columns (Title and Abstract)
abstracts = df[['Title', 'Abstract Note']].dropna()

# Define primary keywords for ethical and legal aspects
primary_keywords = ['ethic', 'ethical', 'legal', 'compliance', 'regulation', 'law', 'governance']

# Define secondary keywords for broader overlaps
secondary_keywords = ['strategy', 'algorithm optimization', 'choice', 'decision-making', 'fairness', 'transparency']

# Filter abstracts containing any primary keywords (case-insensitive)
primary_filtered = abstracts[
    abstracts['Abstract Note'].str.contains('|'.join(primary_keywords), case=False, na=False)
]

# Filter abstracts containing secondary keywords (case-insensitive)
secondary_filtered = abstracts[
    abstracts['Abstract Note'].str.contains('|'.join(secondary_keywords), case=False, na=False)
]

# Combine and remove duplicates (if any)
combined_filtered = pd.concat([primary_filtered, secondary_filtered]).drop_duplicates()

# Display the combined filtered abstracts
print("Abstracts Matching Ethical, Legal, or Broader Decision-Making Keywords:")
display(combined_filtered)

# Optional: Save the results to a CSV file
combined_filtered.to_csv('filtered_ethics_and_decisionmaking_abstracts_ML.csv', index=False)
```

```python
# Install necessary libraries (if not already installed)
# !pip install pandas

# Import libraries
import pandas as pd

# Load the dataset
file_path = '/content/scopus_wire AND arc AND additive AND manufacturing.csv'  # Adjust path for your environment
df = pd.read_csv(file_path)

# Display dataset columns
print("Dataset Columns:")
print(df.columns)

# Check if 'Abstract Note' column exists
if 'Abstract Note' in df.columns:
    # Define primary keywords for ethical and legal aspects
    primary_keywords = ['ethic', 'ethical', 'legal', 'compliance', 'regulation', 'law', 'governance']

    # Define secondary keywords for broader overlaps
    secondary_keywords = ['strategy', 'algorithm optimization', 'choice', 'decision-making', 'fairness', 'transparency']

    # Filter rows containing primary keywords
    primary_filtered = df[
        df['Abstract Note'].str.contains('|'.join(primary_keywords), case=False, na=False)
    ]

    # Filter rows containing secondary keywords
    secondary_filtered = df[
        df['Abstract Note'].str.contains('|'.join(secondary_keywords), case=False, na=False)
    ]

    # Combine both filtered datasets and remove duplicates
    combined_filtered = pd.concat([primary_filtered, secondary_filtered]).drop_duplicates()

    # Display the number of matching rows
    print(f"\nNumber of abstracts matching primary or secondary keywords: {len(combined_filtered)}")

    # Display the filtered dataset
    print("\nFiltered Abstracts Matching Ethical, Legal, or Broader Decision-Making Keywords:")
    print(combined_filtered[['Title', 'Abstract Note']])

    # Save the results to a CSV file
    output_file = 'filtered_ethics_and_decisionmaking_abstracts_ALL_expended.csv'
    combined_filtered.to_csv(output_file, index=False)
    print(f"\nFiltered dataset saved to '{output_file}'")

else:
    raise ValueError("Required column 'Abstract Note' is missing from the dataset.")
```
